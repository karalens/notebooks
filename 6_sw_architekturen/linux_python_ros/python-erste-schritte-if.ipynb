# Logische Abfragen

```python
%matplotlib notebook

import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
dt = 50 / 1000
t_max = 10

x = 0
circle = ax.add_patch(plt.Circle((0, 0.5), 0.1, color='r'))
ax.axis('equal')
ax.set(xlim=[0, 1], ylim=[-0.5, 0.5])

def init():
    circle.set_center([0, 0])
    return

def animate(i):
    global x
    t = i * dt
    x = x + 0.2 * dt
    circle.set_center([x, 0])
    return

ani = matplotlib.animation.FuncAnimation(fig, animate, init_func=init, interval=dt*1000, frames=int(t_max/dt), repeat=True)
ani
```
